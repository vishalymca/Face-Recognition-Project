```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("./group-photo.jpg")
print(img.shape)
print(img)
```

```python
plt.imshow(img)
```

```python
detector = cv2.CascadeClassifier("haarcascade_frontalface_alt (2).xml")
#to detect the faces
```

```python
faces = detector.detectMultiScale(img)
```

```python
faces
```

```python
faces.shape
```

```python
faces[0]
```

```python
Px, Py, w, h = faces[0]  
```

```python
Px #x-Pixel
```

```python
Py #y-Pixel
```

```python
w #width
```

```python
h #height
```

```python
img = cv2.rectangle(img, (Px, Py) , (Px+w, Py+h), (255, 255, 255), 3 )

#img=cv2.rectangle(img,starting pixel of rect,final pixel of rect,
#color of rect,width of border of rectangle)

#Google Search RGB color picker 
#for white color(255,255,255)
```

```python
plt.imshow(img)
```

## plot rect for all faces

```python
for face in faces:
    Px,Py,w,h = face
    img = cv2.rectangle(img, (Px, Py) , (Px+w, Py+h), (255, 255, 255), 3 )
```

```python
plt.imshow(img)
```
